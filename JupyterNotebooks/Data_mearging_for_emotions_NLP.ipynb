```python
def new_name(path:str):
    path_l = path.split('/')
    path_l = path_l[:-1] + ['new_' + path_l[-1]]
    new_path = '/'.join(path_l)
    return new_path  
```

```python
new_name('/content/drive/MyDrive/NLP/go_data.csv')
```

```python
path = '/content/drive/MyDrive/NLP/EMNLP/dialogues_text.txt'
with open(path, 'r') as txtReader:
    with open(new_name(path), 'w') as txtWriter:
        for line in txtReader.readlines():
            line = line.replace('\\n', '')
            txtWriter.write(line)
```

```python
big_df = pd.DataFrame()
for path in path_list:
    # JSON file
    f = open (path, "r")
    print('vwig :'+ path)
    # Reading from file
    data_json = json.loads(f.read())
    df = pd.DataFrame()
    # Iterating through the json
    # list
    for data_list in data_json:
        for data in data_list:
            big_df = big_df.append({'Emotions': data['emotion'],'Text': data['utterance']}, ignore_index=True)
            
    
    # Closing file
    f.close()
```
